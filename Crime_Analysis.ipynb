```python
import matplotlib.pyplot as plt
import pandas as pd

fig = plt.figure(figsize=(15, 10))
# Create bar chart
plt.bar(df["State"], df["Data.Rates.Property.All"], label="Property Crime")
plt.bar(df["State"], df["Data.Rates.Violent.All"], label="Violent Crime")
plt.xticks(rotation=90)
plt.xlabel("State")
plt.ylabel("Crime Rate per 100,000 Population")
plt.title("Comparison of Crime Rates between States")
plt.legend()
plt.show()
```

```python
sns.set(style="whitegrid")

# Filter data for murder and rape rates
mr_vars = ["Data.Rates.Violent.Murder", "Data.Rates.Violent.Rape"]
mr_data = crime_data1[["State"] + mr_vars]

# Calculate total number of murders and rapes per state
mr_data["Total"] = mr_data[mr_vars[0]] + mr_data[mr_vars[1]]
mr_data = mr_data.groupby("State").sum().sort_values("Total", ascending=False)

# Select top 5 states
top_5_states = mr_data.head(5).index

# Filter data for top 5 states
top_5_data = crime_data1.loc[crime_data1["State"].isin(top_5_states), ["State"] + mr_vars]

# Create boxplot
fig, ax = plt.subplots(figsize=(10, 6))
sns.boxplot(x="State", y="value", hue="variable", data=pd.melt(top_5_data, id_vars=["State"]), ax=ax)
ax.set_title("Murder and Rape Rates for Top 5 States")
plt.show()
```

```python
import seaborn as sns

fig = plt.figure(figsize=(12, 10))

# select relevant columns
crime_rates = df[['Data.Rates.Property.All', 'Data.Rates.Violent.All']]

# plot violin plot
sns.violinplot(data=crime_rates, inner='stick')
plt.ylabel('Crime Rates')
plt.xticks([0, 1], ['Property', 'Violent'])
plt.title('Comparison of Property and Violent Crime Rates')
plt.show()
```

```python
# Select the columns needed for the bubble chart
bubble_data = df[['State', 'Data.Population', 'Data.Rates.Property.All', 'Data.Rates.Violent.All']]

# Create the scatter plot with marker size representing population
plt.scatter(x=bubble_data['Data.Rates.Property.All'], 
            y=bubble_data['Data.Rates.Violent.All'], 
            s=bubble_data['Data.Population']/1000, 
            alpha=0.5)

# Add labels to the markers
for i, row in bubble_data.iterrows():
    plt.text(row['Data.Rates.Property.All'], 
             row['Data.Rates.Violent.All'], 
             row['State'], 
             fontsize=8,
             horizontalalignment='center', 
             verticalalignment='center')

# Add axis labels and title
plt.xlabel('Property Crime Rate per 100,000 People')
plt.ylabel('Violent Crime Rate per 100,000 People')
plt.title('Crime Rates vs. Population in Different States')

# Add a legend for the marker size
plt.legend(['Population (in thousands)'], loc='upper left')

# Show the plot
plt.show()
```

```python
import pandas as pd
import plotly.graph_objects as go
df_state = crime_data1.groupby("State").sum().reset_index()

# Create a packed bubble chart
fig = go.Figure(data=go.Scatter(
    x=df_state["State"],
    y=df_state["Data.Rates.Violent.Murder"],
    mode='markers',
    marker=dict(
        size=df_state["Data.Rates.Violent.All"],
        sizemode='diameter',
        sizeref=0.85,
        sizemin=3,
        color=df_state["Data.Rates.Violent.Murder"],
        colorbar=dict(title="Murder"),
        colorscale="Reds",
        opacity=0.7),
    text=df_state["State"]
))

# Update the layout of the chart
fig.update_layout(
    title="State Crime Statistics",
    xaxis_title="State",
    yaxis_title="Murder",
    plot_bgcolor="white"
)

# Display the chart
fig.show()
```
